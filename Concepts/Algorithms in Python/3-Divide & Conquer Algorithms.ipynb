# Merge Sort

```python
def iterative_merge_sort(left, right):
    temp = []
    while min(len(left), len(right)) > 0:
        if left[0] > right[0]:
            insert = right.pop(0)
            temp.append(insert)
        elif left[0] <= right[0]:
            insert = left.pop(0)
            temp.append(insert)
    if len(left) > 0:
        for i in left:
            temp.append(i)
    if len(right) > 0:
        for i in right:
            temp.append(i)
    return temp

left = [2,5,6,10]
right = [3,4,12,20]
print(iterative_merge_sort(left,right))
```

```python
def recursive_merge_sort(arr):
    if len(arr) <= 1:
        return arr
    middle = len(arr) // 2
    left = recursive_merge_sort(arr[:middle])
    right = recursive_merge_sort(arr[middle:])
    merged = []
    while left and right:
        if left[0] <= right[0]:
            merged.append(left.pop(0))
        else:
            merged.append(right.pop(0))
    merged.extend(right if right else left)
    return merged

array = [29,3,7,4,8,5,0]
print(recursive_merge_sort(array))           
```

```python
def merge_sort(arr):
    mid = len(arr) // 2
    left = sorted(arr[:mid])
    right = sorted(arr[mid:])
    temp = []
    while min(len(left), len(right)) > 0:
        if left[0] > right[0]:
            insert = right.pop(0)
            temp.append(insert)
        elif left[0] <= right[0]:
            insert = left.pop(0)
            temp.append(insert)
    if len(left) > 0:
        for i in left:
            temp.append(i)
    if len(right) > 0:
        for i in right:
            temp.append(i)
    return temp

array = [29,3,7,4,8,5,0]
print(merge_sort(array)) 
```

# Strassen Algorithm

```python
import numpy as np

def strass_itr(x, y):
    a,b,c,d = x[0,0],x[0,1],x[1,0],x[1,1]
    e,f,g,h = y[0,0],y[0,1],y[1,0],y[1,1]
    
    p1 = a*(f-h)
    p2 = (a+b)*h
    p3 = (c+d)*e
    p4 = d*(g-e)
    p5 = (a+d)*(e+h)
    p6 = (b-d)*(g+h)
    p7 = (a-c)*(e+f)
    
    c1 = (p5+p4-p2+p6)
    c2 = (p1+p2)
    c3 = (p3+p4)
    c4 = (p1+p5-p3-p7)
    
    return np.array([[c1,c2],[c3,c4]])

x = np.array([[1,2],[2,3]])
y = np.array([[2,3],[3,4]])
print(strass_itr(x,y))
```

```python
import numpy as np

x = np.array([[1,2],[2,3]])
y = np.array([[2,3],[3,4]])

def split(matrix):
    row, col = matrix.shape
    row2, col2 = row // 2, col // 2
    return matrix[:row2, :col2], matrix[:row2, col2:], matrix[row2:, :col2], matrix[row2:, col2:]

def strass_recur(x,y):
    if len(x) == 1 and len(y) == 1:
        # Base case: if the matrices are 1x1, just multiply them
        return np.array([[x[0][0] * y[0][0]]])

    else:
        # Split the matrices into quarters and recursively compute the products
        a,b,c,d = split(x)
        e,f,g,h = split(y)
        
        p1 = strass_recur(a, f-h)
        p2 = strass_recur(a+b, h)
        p3 = strass_recur(c+d, e)
        p4 = strass_recur(d, g-e)
        p5 = strass_recur(a+d, e+h)
        p6 = strass_recur(b-d, g+h)
        p7 = strass_recur(a-c, e+f)
        
        # Combine the products to form the final matrix
        c1 = p5 + p4 - p2 + p6
        c2 = p1 + p2
        c3 = p3 + p4
        c4 = p1 + p5 - p3 - p7
        
        result = np.vstack((np.hstack((c1,c2)), np.hstack((c3,c4))))
        return result

print(strass_recur(x, y))
```

```python
import numpy as np

def strassen_iterative(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]

    # Split matrices into quadrants
    for k in range(0, n, 2):
        for l in range(0, n, 2):
            A11, A12, A21, A22 = A[k][l], A[k][l+1], A[k+1][l], A[k+1][l+1]
            B11, B12, B21, B22 = B[k][l], B[k][l+1], B[k+1][l], B[k+1][l+1]

            # Compute 7 products iteratively
            M1 = (A11 + A22) * (B11 + B22)
            M2 = (A21 + A22) * B11
            M3 = A11 * (B12 - B22)
            M4 = A22 * (B21 - B11)
            M5 = (A11 + A12) * B22
            M6 = (A21 - A11) * (B11 + B12)
            M7 = (A12 - A22) * (B21 + B22)

            # Compute result quadrant
            C11 = M1 + M4 - M5 + M7
            C12 = M3 + M5
            C21 = M2 + M4
            C22 = M1 - M2 + M3 + M6

            # Combine quadrants into result matrix
            C[k][l], C[k][l+1], C[k+1][l], C[k+1][l+1] = C11, C12, C21, C22
    return C

# Example usage
A = [[1, 2], [2, 3]]
B = [[2, 3], [3, 4]]
C = strassen_iterative(A, B)

print(np.array(C))
```

```python
import numpy as np

def strassen_recursive(matrix1, matrix2):
    """
    Multiplies two matrices using the Strassen Algorithm.
    Assumes that the matrices are square and have dimensions that are powers of 2.
    """
    # Base case: if the matrices are 1x1, just multiply them
    if len(matrix1) == 1:
        return matrix1 * matrix2
    
    # Split the matrices into quarters. This will be done recursively.
    size = len(matrix1)
    split_size = size // 2
    a = matrix1[:split_size, :split_size]
    b = matrix1[:split_size, split_size:]
    c = matrix1[split_size:, :split_size]
    d = matrix1[split_size:, split_size:]
    e = matrix2[:split_size, :split_size]
    f = matrix2[:split_size, split_size:]
    g = matrix2[split_size:, :split_size]
    h = matrix2[split_size:, split_size:]
    
    # Compute the seven products recursively
    p1 = strassen_recursive(a, f - h)
    p2 = strassen_recursive(a + b, h)
    p3 = strassen_recursive(c + d, e)
    p4 = strassen_recursive(d, g - e)
    p5 = strassen_recursive(a + d, e + h)
    p6 = strassen_recursive(b - d, g + h)
    p7 = strassen_recursive(a - c, e + f)
    
    # Compute the four quadrants of the result matrix
    top_left = p5 + p4 - p2 + p6
    top_right = p1 + p2
    bottom_left = p3 + p4
    bottom_right = p1 + p5 - p3 - p7
    
    # Combine the quadrants into a single matrix and return it
    result = np.vstack((np.hstack((top_left, top_right)), np.hstack((bottom_left, bottom_right))))
    return result

matrix1 = np.array([[1, 2], [2, 3]])
matrix2 = np.array([[2, 3], [3, 4]])
result = strassen_recursive(matrix1, matrix2)
print(result)
```
